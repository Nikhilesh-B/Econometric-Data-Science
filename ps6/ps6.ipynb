**1a)**

**Forward direction (if $X_i$ is exogenous then $Z_i$ is a valid instrument).**  
For a valid instrument we need two conditions to be satisfied:  
1. Relevance: $\operatorname{Cov}(Z_i, X_i) \neq 0$  
2. Exogeneity: $\operatorname{Cov}(e_i, Z_i) = 0$

By definition, condition (1) holds. Condition (2) also holds because if $X_i$ is exogenous then  
$\operatorname{Cov}(X_i, e_i) = 0$.  
Since $X_i = Z_i + \varepsilon_i$, this implies  
$\operatorname{Cov}(Z_i + \varepsilon_i, e_i) = 0$,  
which expands to  
$\operatorname{Cov}(Z_i, e_i) + \operatorname{Cov}(\varepsilon_i, e_i) = 0$.  
Because $\varepsilon_i$ is randomly generated noise, $\operatorname{Cov}(\varepsilon_i, e_i) = 0$.  
Therefore, $\operatorname{Cov}(Z_i, e_i) = 0$, establishing instrument exogeneity.

---

**Backward direction (if $Z_i$ is a valid instrument then $X_i$ is exogenous).**  
Since $Z_i$ is a valid instrument, we know  
$\operatorname{Cov}(e_i, Z_i) = 0$.  
Using $X_i = Z_i + \varepsilon_i$, this implies  
$\operatorname{Cov}(e_i, X_i - \varepsilon_i) = 0$,  
which expands to  
$\operatorname{Cov}(e_i, X_i) - \operatorname{Cov}(e_i, \varepsilon_i) = 0$.  
Because $\varepsilon_i$ is completely random, $\operatorname{Cov}(e_i, \varepsilon_i) = 0$.  
Thus, $\operatorname{Cov}(e_i, X_i) = 0$, meaning $X_i$ is exogenous.

**1b)**

**Calculate the asymptotic variance of the IV and compare it to the OLS. Which estimator would you prefer?**

**1. Derivation of Asymptotic Variance of IV**

Let the model be $Y_i = \alpha + \beta X_i + e_i$. The IV estimator using instrument $Z_i$ is given by:
$$ \hat{\beta}_{IV} = \frac{\sum (Z_i - \bar{Z})(Y_i - \bar{Y})}{\sum (Z_i - \bar{Z})(X_i - \bar{X})} $$

Substituting $Y_i = \alpha + \beta X_i + e_i$ into the estimator:
$$ \hat{\beta}_{IV} = \beta + \frac{\sum (Z_i - \bar{Z})e_i}{\sum (Z_i - \bar{Z})(X_i - \bar{X})} $$

Subtracting $\beta$ from both sides and scaling by $\sqrt{n}$:
$$ \sqrt{n}(\hat{\beta}_{IV} - \beta) = \frac{\frac{1}{\sqrt{n}}\sum (Z_i - \bar{Z})e_i}{\frac{1}{n}\sum (Z_i - \bar{Z})(X_i - \bar{X})} $$

By the Central Limit Theorem (CLT) and Law of Large Numbers (LLN) as $n \to \infty$:
*   **Numerator:** $\frac{1}{\sqrt{n}}\sum (Z_i - \bar{Z})e_i \xrightarrow{d} N(0, Var(Z_i e_i))$. Under homoskedasticity, $Var(Z_i e_i) = \sigma_e^2 Var(Z)$.
*   **Denominator:** $\frac{1}{n}\sum (Z_i - \bar{Z})(X_i - \bar{X}) \xrightarrow{p} Cov(Z, X)$.

Thus, the asymptotic variance is:
$$ Avar(\hat{\beta}_{IV}) = \frac{\sigma_e^2 \cdot Var(Z)}{Cov(X, Z)^2} $$

Using the definition of correlation $\rho_{XZ} = \frac{Cov(X, Z)}{\sqrt{Var(X)Var(Z)}}$, we can rewrite $Cov(X, Z)^2 = \rho_{XZ}^2 \cdot Var(X) \cdot Var(Z)$. Substituting this back:

$$ Avar(\hat{\beta}_{IV}) = \frac{\sigma_e^2 \cdot Var(Z)}{\rho_{XZ}^2 \cdot Var(X) \cdot Var(Z)} = \frac{\sigma_e^2}{Var(X) \cdot \rho_{XZ}^2} $$

**2. Asymptotic Variance of OLS**

The asymptotic variance of the OLS estimator is:
$$ Avar(\hat{\beta}_{OLS}) = \frac{\sigma_e^2}{Var(X)} $$

**3. Comparison and Conclusion**

Comparing the two variances:
$$ Avar(\hat{\beta}_{IV}) = \frac{1}{\rho_{XZ}^2} \cdot Avar(\hat{\beta}_{OLS}) $$

Since the correlation squared $\rho_{XZ}^2 \leq 1$ (and strictly less than 1 unless X and Z are perfectly correlated), we have:
$$ Avar(\hat{\beta}_{IV}) \geq Avar(\hat{\beta}_{OLS}) $$

**Conclusion:**
If $X_i$ is exogenous, both OLS and IV are consistent estimators. However, OLS is **more efficient** (has smaller variance) than IV. Therefore, if $X_i$ is exogenous, we would strictly **prefer the OLS estimator**.

**1c)**

**The IV estimator of the returns to education ($\beta$) ... with a dummy instrument $Z_i$ ... is exactly the same as the Wald estimator. Show this.**

The IV estimator is given by:
$$ \hat{\beta}_{IV} = \frac{\text{Cov}(Y_i, Z_i)}{\text{Cov}(X_i, Z_i)} $$

Since $Z_i$ is a binary dummy variable taking values 0 and 1, the covariance can be rewritten in terms of group means.

**1. Denominator: Cov(X, Z)**
$$ \text{Cov}(X_i, Z_i) = E[X_i Z_i] - E[X_i]E[Z_i] $$
Let $p = P(Z_i = 1)$. Then $E[Z_i] = p$.
$E[X_i Z_i] = p \cdot E[X_i | Z_i=1] + (1-p) \cdot 0 = p \cdot \bar{x}_1$.
$E[X_i] = p \cdot \bar{x}_1 + (1-p) \cdot \bar{x}_0$.

So,
$$ \text{Cov}(X_i, Z_i) = p \bar{x}_1 - (p \bar{x}_1 + (1-p)\bar{x}_0)p $$
$$ = p \bar{x}_1 - p^2 \bar{x}_1 - p(1-p)\bar{x}_0 $$
$$ = p(1-p)\bar{x}_1 - p(1-p)\bar{x}_0 $$
$$ = p(1-p)(\bar{x}_1 - \bar{x}_0) $$

**2. Numerator: Cov(Y, Z)**
Similarly, for $Y$:
$$ \text{Cov}(Y_i, Z_i) = p(1-p)(\bar{y}_1 - \bar{y}_0) $$
Where $\bar{y}_1 = \overline{wage}_1$ and $\bar{y}_0 = \overline{wage}_0$.

**3. The Ratio**
$$ \hat{\beta}_{IV} = \frac{p(1-p)(\overline{wage}_1 - \overline{wage}_0)}{p(1-p)(\overline{educ}_1 - \overline{educ}_0)} $$
$$ \hat{\beta}_{IV} = \frac{\overline{wage}_1 - \overline{wage}_0}{\overline{educ}_1 - \overline{educ}_0} $$

This confirms that with a binary instrument, the IV estimator is indeed the Wald estimator (the ratio of the difference in means).

## a) 

Estimate the price elasticity of demand by using OLS to regress the log
of the quantity of grain shipped on the log of the price and the full set of
monthly binary indicators. (Since monthly binary indicators are named
seas1 ,..., seas13 in this dataset, in Stata you can include them in the
regression using seas* as a shorthand.) What is the estimated value of the
demand elasticity and its standard error?


```python
df['l_quantity'] = np.log(df['quantity'])
df['l_price'] = np.log(df['price'])
```

```python
features = ['l_price']+['seas'+str(i) for i in range(1,13)]
features_df = df[features]

model = sm.OLS(endog=df['l_quantity'], exog=features_df).fit(cov_type='HC0')
print(model.summary())
```

```python
print(f"Estimated demand elasticity: {model.params.iloc[0]}%")
print(f"Estimated demand elasticity standard error: {model.bse.iloc[0]}")
```

## (e) Run 3 different TSLS regressions

```python

# Define variables
# Using the same seasonal indicators as in the OLS (seas1 to seas12)
# Note: linearmodels requires us to be explicit about the constant/exogenous vars.
# Since the OLS did not use an explicit constant but used all seasonal dummies (assuming 12 months), 
# we treat them as the exogenous controls included in both stages.

# Check if seas13 exists to be safe, otherwise use 1-12 as user did
seas_cols = [col for col in df.columns if 'seas' in col]
# Based on user's previous cell, they used range(1,13)
exog_data = df[[f'seas{i}' for i in range(1, 13)]]

# Dependent and Endogenous
dep = df['l_quantity']
endog = df['l_price']

# (i) IV: cartel
print("--- Regression (i): IV = cartel ---")
iv1 = IV2SLS(dep, exog_data, endog, df[['cartel']]).fit(cov_type='robust')
print(iv1.summary)

# (ii) IV: ice
print("\n--- Regression (ii): IV = ice ---")
iv2 = IV2SLS(dep, exog_data, endog, df[['ice']]).fit(cov_type='robust')
print(iv2.summary)

# (iii) IV: cartel and ice
print("\n--- Regression (iii): IV = cartel + ice ---")
iv3 = IV2SLS(dep, exog_data, endog, df[['cartel', 'ice']]).fit(cov_type='robust')
print(iv3.summary)
```

```python
print("\n" + "="*50)
print("J-TEST STATISTIC FOR REGRESSION (iii)")
print("="*50)
print(iv3.sargan)

print("\n" + "="*50)
print("FIRST STAGE DIAGNOSTICS (F-STATISTICS)")
print("="*50)
print("--- Regression (i) ---")
print(iv1.first_stage.diagnostics)
print("\n--- Regression (ii) ---")
print(iv2.first_stage.diagnostics)
print("\n--- Regression (iii) ---")
print(iv3.first_stage.diagnostics)
```

## (g) Perform J-test of overidentifying restrictions in regression (e) (iii). What conclusions can you draw?

The J-test (Sargan-Hansen test) tests the null hypothesis that the instruments are valid (uncorrelated with the error term), assuming at least one instrument is valid.

*   **Null Hypothesis ($H_0$):** All instruments are valid (exogenous).
*   **Alternative Hypothesis ($H_A$):** At least one instrument is invalid.

Look at the **"Sargan"** or **"J-statistic"** p-value in the summary for regression (iii) or in the explicit output below it.
*   If p-value < 0.05, we reject the null hypothesis (evidence that instruments are invalid).
*   If p-value > 0.05, we fail to reject the null hypothesis (instruments seem valid).


The p-value is far less than 0.05 hence at least one of the instruments seems to be invalid.
